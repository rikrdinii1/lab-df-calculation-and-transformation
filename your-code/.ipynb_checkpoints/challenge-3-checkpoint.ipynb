# Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

Next, import `Orders.csv` from the "subsetting" lab folder into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# enter your code here
orders = pd.read_csv("Orders.csv")
orders.head()
```

"Identify VIP and Preferred Customers" is the non-technical goal of your boss. You need to translate that goal into technical languages that data analysts use:

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
orders.describe([.25, .75, .95])
orders.head()
```

```python
# obtenemos el total comprado por cada uno de los clientes
tot_amount_spend = orders.groupby('CustomerID').agg({'amount_spent':'sum'})
tot_amount_spend.head()
# obtenemos las estadisticas
quantile75 = float(tot_amount_spend.quantile(.75))
quantile95 = float(tot_amount_spend.quantile(.95))

print(quantile75,quantile95)
```

```python
#ponemos las etiquetas
tot_amount_spend.head(2)
```

```python
# Determinamos si el tipo de cliente
tot_amount_spend['Mayor Q95'] = tot_amount_spend['amount_spent'].apply(lambda x: 1 if x >=quantile95 else 0)
tot_amount_spend['Entre Q95 y Q75'] = tot_amount_spend['amount_spent'].apply(lambda x: 1 if (x <quantile95)&(x >=quantile75) else 0)
tot_amount_spend['Menor Q75'] = tot_amount_spend['amount_spent'].apply(lambda x: 1 if (x <quantile75) else 0)
```

```python
tot_amount_spend.head(2)
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

# Q2: How to identify which country has the most VIP Customers?

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

Provide your solution for Q2 below:

```python
# your code here
# Primero unimos ambas tablas
tot_amount_spend_quantiles = pd.merge(tot_amount_spend,orders,on='CustomerID')
tot_amount_spend_quantiles.head(2)

```

```python
#Obtenemos el total por cada uno de los paises
tot_pais = tot_amount_spend_quantiles.groupby('Country').agg({'Mayor Q95':'sum','Entre Q95 y Q75':'sum'})
tot_pais.sort_values(by='Mayor Q95',ascending=False).head(10)
# Con esto obtenemos los 10 que tienen mas clientes por encima del q95
```

```python
#Obtenemos los paises con mas clientes combinados
tot_pais['VIP + Prefered']= tot_pais['Mayor Q95'] + tot_pais['Entre Q95 y Q75']
tot_pais.sort_values(by='VIP + Prefered',ascending=False).head(10)
```
